```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import datetime

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
print(len(cities))
print(cities[:2])
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
def print_log(str,fp):
    print(str)
    print(str,file=fp)
```

```python
base_url = "http://api.openweathermap.org/data/2.5/weather?q={}&appid={}&units=imperial"
batch_size = 50
set_num = 0
success = 0
failure = 0
output_log_file = "../Output/Log/execution_log.log"

with open(output_log_file, "w") as opfile:
    weather_list=[]
    print_log("-" * 30, opfile)
    print_log("Beginning Data Retrieval", opfile)
    print_log("-" * 30, opfile)

    for id, city in enumerate(cities, start=0):
        if id % batch_size == 0:
            set_num += 1
            #time.sleep(60) #If a single batch is processed then wait for a minute before starting processing a new batch
        
        url=base_url.format(city,api_key)
        try:
            print_log(f"Processing Record {id % batch_size} of Set {set_num} | {city}", opfile)
            response=requests.get(url)

            if response.status_code != 200:
                print_log("City not found. Skipping...", opfile)
                failure += 1
            else:
                success +=1
                resp_json = response.json()
                weather_list.append([city, 
                                     resp_json['clouds']['all'], 
                                     resp_json['sys']['country'],
                                     resp_json['dt'],
                                     resp_json['main']['humidity'],
                                     resp_json['coord']['lat'],
                                     resp_json['coord']['lon'],
                                     resp_json['main']['temp_max'],
                                     resp_json['wind']['speed']
                                    ])
        except Exception as e:
            print_log(e, opfile)
    print_log("-" * 30, opfile)
    print_log("Data Retrieval Complete", opfile)
    print_log("-" * 30, opfile)
    
    print_log("\n\n",opfile)
    print_log("-" * 30, opfile)
    print_log("Summary Statistics", opfile)
    print_log("-" * 30, opfile)
    print_log(f"{'Total Rows Processed':25s}" + f": " + str(len(cities)), opfile)
    print_log(f"{'Total Rows Successful':25s}" + f": " + str(success), opfile)
    print_log(f"{'Total Rows Failed':25s}"+ f": "+ str(failure), opfile)
    print_log("-" * 30, opfile)
```

```python
print(len(weather_list))
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weather_df=pd.DataFrame(weather_list,columns=["City","Cloudiness","Country","Date","Humidity","Lat", "Lng", "Max Temp", "Wind Speed"])
weather_df.to_csv("../Output/Data/Weather_DataFrame.csv")
```
